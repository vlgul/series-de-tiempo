## **Modelos de regresión**

- 1) Capturar datos
- 2) Ajustar el modelo
- 3) Estimaciones puntuales para t=25, 26, 27
- 4) Estimaciones por intervalos de confianza para t= 25, 26, 27

```r
x<-seq(1,24)
x2<-(seq(1,24)^2)
y<-c(297,249,340,406,464,481,549,553,556,642,670,712,808,809,867,855,965,921,956,990,1019,1021,1033,1127)
matriz<-matrix(c(x,x2,y), ncol=3, byrow=F)
matriz
```

Modelar en una regresión lineal

```r
f <- lm(y ~ x + x2)
summary(f)
```

```r
plot(f)
```

```r
var<-var(y)
S<-31.25
```

```r
f
f$coefficients[1]
f$coefficients[2]
f$coefficients[3]
```

Despues de extraer los datos el valor de S y Yt para calcular los intervalos de confianza con la funcion summary(). Valor para Yt cuando t=25

```r
t=25
A= f$coefficients[1]+((f$coefficients[2])*t)-f$coefficients[3]*t*t
cat("Valor de Yt cuando t=25: ",A)
```

Valor para Yt cuando t=26

```r
t1=26
B= f$coefficients[1]+((f$coefficients[2])*t1)-f$coefficients[3]*t1*t1
cat("Valor de Yt cuando t=26: ",B)
```

Valor para Yt cuando t=27

```r
t2=27
C=f$coefficients[1]+((f$coefficients[2])*t2)-f$coefficients[3]*t2*t2
cat("Valor de Yt cuando t=27: ",C)
```

Intervalos de confianza Para t=25 el valor teorico alfa=0.025 y g.l. = n-1 = 25-1= 24

```r
t01<-2.07
w=1+1/t
p<-var*(t-1)
j=sum(y-mean(y))^2
SU1=A+t01*S*sqrt(w+j/p)
INF1=A-t01*S*sqrt(w+j/p)
cat("Valor Inferior",INF1,"\n")
cat("Valor Superior",SU1)
```

Para t=26 el valor teorico alfa=0.025 y g.l. = n-1 = 26-1= 25

```r
t02<-2.06
y1=1+1/t1
q<-var*(t1-1)
k=sum(y-mean(y))^2
SU2=B+t02*S*sqrt(y1+k/q)
INF2=B-t02*S*sqrt(y1+k/q)
cat("Valor Inferior",INF2,"\n")
cat("Valor Superior",SU2)
```

Para t=27 el valor teorico alfa=0.025 y g.l. = n-1 = 27-1= 26

```r
t03<-2.055
z=1+1/t2
r<-var*(t2-1)
l=sum(y-mean(y))^2
SU3=C+t03*S*sqrt(z+l/r)
INF3=C-t03*S*sqrt(z+l/r)
cat("Valor Inferior",INF3,"\n")
cat("Valor Superior",SU3)
```
